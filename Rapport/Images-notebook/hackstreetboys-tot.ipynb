**Prototype**

In the prototype phase, we started developing a simple map prototype in Streamlit. We used unsupervised machine learning specifically KMeans clustering to group playgrounds based on their proximity to public transport stops.

```python
# We used parts of this code to display a prototype map in Streamlit where playgrounds are color-coded based on their distance to 
# public transport stops. The KMeans algorithm groups them into four clusters, which are then visualized with different
# marker colors. 
 

# Beräkna avstånd till närmaste hållplats
lekplats_df["avstånd_hållplats"] = lekplats_df.apply(
    lambda row: närmaste_avstånd(row["lat"], row["lon"], hållplatser_df), axis=1
)

# Skala datan och klustra med KMeans
scaler = StandardScaler()
scaled = scaler.fit_transform(lekplats_df[["avstånd_hållplats"]])
kmeans = KMeans(n_clusters=4, random_state=42)
lekplats_df["cluster"] = kmeans.fit_predict(scaled)

# Färger per kluster
colors = ['green', 'orange', 'red', 'purple']
lekplats_df["color"] = lekplats_df["cluster"].map(lambda x: colors[x])

# Skapa karta
map = folium.Map(location=[57.7, 11.97], zoom_start=12)
for _, row in lekplats_df.iterrows():
    folium.Marker(
        location=[row['lat'], row['lon']],
        icon=folium.Icon(color=row['color']),
        popup=row['namn']
    ).add_to(map)

folium_static(map)
```

### Data preparation and analysis

At the beginning of the project, we reviewed the dataset from Trafiklab (KoDa) and investigated other data sources that could be linked to our problem definition: to identify child-friendly activities in the Gothenburg area. Although we were initially unsure of exactly which dataset we needed, we knew that we wanted to focus on playgrounds.

We found a relevant dataset from OpenStreetMap (OSM) with information about all playgrounds in Gothenburg, which we extracted using the keyword "playground". To make the data more manageable and focused, we chose to filter both the OSM and KoDa data geographically to a specific coordinate area within Gothenburg. This enabled more efficient handling and processing at an early stage of development.

During an initial review of the playground data, we discovered that some playgrounds had the attribute "access: private". By verifying these via Google Maps, we were able to establish that they were located within private residential areas and were not accessible to the public. Therefore, we chose to filter out these records to ensure that our analysis only included publicly accessible playgrounds.

To get a clear overview, we used a map where we plotted the coordinates of the stops and playgrounds. This helped us to visually understand how the playgrounds are distributed in relation to public transport. We then applied AI-based cluster analysis using the K-means algorithm to identify patterns in the data. To identify the appropriate number of clusters (k) for the analysis of the distance of playgrounds to the nearest stop, a cluster analysis using K-means was performed. As a first step, different values of k in the range of 2 to 6 were tested. For each value, both inertia and silhouette scores were calculated.

The purpose of this process was to identify a balance between the compactness and separation of the clusters. The inertia value was visualized in an Elbow Plot, where the goal was to identify a clear “elbow point” where further increases in k do not lead to any major improvement in the compactness of the clusters. At the same time, a high silhouette value confirms that the clusters are well separated and meaningful.

Based on these graphs, k = 4 was chosen as the most appropriate value for further analysis, as this offered a good balance between low inertia and high silhouette score. Through this analysis, we were able to group playgrounds based on their distance to the nearest bus stop.

We used an elbow plot to determine the optimal number of clusters, which improved the accuracy of the model. Furthermore, we used visualizations to summarize the content of each cluster, for example how many playgrounds were in each group and their geographical distribution.

To increase the stability of the model, we introduced standardization before clustering, which gave more reliable results. Finally, we implemented an interactive prototype in Streamlit that was integrated with our GitHub repo, which enables version control and further development in collaboration within the team.

Through this iterative process, we have been able to gradually define and process relevant data sources, develop a functioning analysis model, and draw insights that directly influenced the design of our prototype.
